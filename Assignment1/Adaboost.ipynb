```python
#imports
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import copy


from google.colab import drive

from sklearn import tree
from sklearn.base import clone
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import plot_tree
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
```

```python
drive.mount('/content/drive')
file_path = '/content/drive/MyDrive/Colab Notebooks/SENG 474/Assignment1/spambase_augmented.csv'
df = pd.read_csv(file_path)
```

```python
print(df.head())
```

```python
x = df.iloc[:,:-1] #features
y = df.iloc[:,-1] #label/last col
```

```python
print(df.head())
```

```python
print(np.shape(df))
print(x.shape)
print(y.shape)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2, random_state=0, shuffle=False)
```

```python
dtc = DecisionTreeClassifier(max_depth=3)
dtc = dtc.fit(x_train, y_train)
```

```python
abc = AdaBoostClassifier(estimator=dtc, n_estimators=56, random_state = 0, learning_rate = 1)
abc = abc.fit(x_train, y_train)
```

```python
print(abc.score(x_test, y_test))
```

```python
#Estimator
estimator_dict = {"EstimatorNumber":[1, 10, 20, 50],"TrainLoss":[],"TestLoss":[]}
for i in estimator_dict['EstimatorNumber']:

  abc = AdaBoostClassifier(n_estimators=i, random_state = 0, learning_rate = 1)
  abc.fit(x_train, y_train)
  estimator_dict["EstimatorNumber"].append(i)
  estimator_dict["TrainLoss"].append(1-(abc.score(x_train, y_train)))
  estimator_dict["TestLoss"].append(1-(abc.score(x_test, y_test)))
```

```python
plt.figure().set_figwidth(15)
plt.grid(True)

plt.xlabel("Loss")
plt.ylabel("Training iterations")
plt.title("Loss & Train Size")

x_values = range(len(estimator_dict['EstimatorNumber']))

plt.plot(x_values, estimator_dict['TrainLoss'], label="Train Loss")
plt.plot(x_values, estimator_dict['TestLoss'], label="Test Loss")

plt.legend()
plt.show()
```

```python
#Learning rate
estimator_dict = {"LearningRate":[0.0001, 0.001, 0.005, 0.01, 0.1, 1, 10, 100],"TrainLoss":[],"TestLoss":[]}
for i in estimator_dict['LearningRate']:

  abc = AdaBoostClassifier(n_estimators=50, random_state = 0, learning_rate = i)
  abc.fit(x_train, y_train)
  estimator_dict["LearningRate"].append(i)
  estimator_dict["TrainLoss"].append(1-(abc.score(x_train, y_train)))
  estimator_dict["TestLoss"].append(1-(abc.score(x_test, y_test)))
```

```python
plt.figure().set_figwidth(15)
plt.grid(True)

plt.xlabel("Loss")
plt.ylabel("Training iterations")
plt.title("Loss & Train Size")

x_values = range(len(estimator_dict['LearningRate']))

plt.plot(x_values, estimator_dict['TrainLoss'], label="Train Loss")
plt.plot(x_values, estimator_dict['TestLoss'], label="Test Loss")

plt.legend()
plt.show()
```
